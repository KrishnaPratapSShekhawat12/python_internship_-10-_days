````python
balance = 5000
pin = 1234
count = 0
MAX_PIN_ATTEMPTS = 3

while True:
    pin_input = input(f"Enter your PIN (you have {MAX_PIN_ATTEMPTS - count} attempts left, or 'exit' to quit): ").strip().lower()

    if pin_input == 'exit':
        print("Exiting ATM. Goodbye!")
        break

    try:
        user_pin = int(pin_input)
    except ValueError:
        print("Invalid input. Please enter a numerical PIN or 'exit'.")
        if count == MAX_PIN_ATTEMPTS - 1:
            print(f"Too many invalid PIN attempts. Exiting after {MAX_PIN_ATTEMPTS} tries.")
            break
        count += 1
        continue

    if user_pin == pin:
        count = 0
        while True:
            print(f"\nYour current balance is: {balance}")
            try:
                choose = int(input('''1. Withdraw
2. Add money
3. Exit transactions
Enter your choice: '''))
            except ValueError:
                print("Invalid input. Please enter a number for your choice.")
                continue

            if choose == 1:
                print("Current balance : ", balance)
                try:
                    amount = int(input("Enter the amount to withdraw: "))
                except ValueError:
                    print("Invalid input. Please enter a numerical amount.")
                    continue

                if amount <= 0:
                    print("Withdrawal amount must be positive.")
                    continue
                elif balance >= amount:
                    balance -= amount
                    print(f"Amount Withdrawn : {amount}")
                    print(f"Current balance : {balance}")
                else:
                    print(f"Insufficient balance. Your current balance is: {balance}")
            elif choose == 2:
                print("Current balance : ", balance)
                try:
                    amount = int(input("Enter the amount to add: "))
                except ValueError:
                    print("Invalid input. Please enter a numerical amount.")
                    continue

                if amount > 0:
                    balance += amount
                    print(f"Amount added : {amount}")
                    print(f"Current balance : {balance}")
                else:
                    print("Invalid amount. Please enter a positive value.")
            elif choose == 3:
                print("Exiting transactions. Please enter your PIN again to continue or 'exit'.")
                break
            else:
                print("Invalid choice. Please select 1, 2 or 3.")
    else:
        print("Wrong pin entered.")
        if count == MAX_PIN_ATTEMPTS - 1:
            print(f"Too many wrong PIN attempts ({MAX_PIN_ATTEMPTS} tries). Card locked. Exiting.")
            break
        count += 1
````
