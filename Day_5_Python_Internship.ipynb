```python
from google.colab import output
from base64 import b64decode

audio = output.eval_js('''
async function record() {
  const s = await navigator.mediaDevices.getUserMedia({audio:true});
    const r = new MediaRecorder(s); let d=[];
      r.ondataavailable=e=>d.push(e.data); r.start();
        await new Promise(res=>setTimeout(res,3000)); r.stop();
          await new Promise(res=>r.onstop=res);
            let b=new Blob(d,{type:'audio/wav'});
              let r2=new FileReader(); r2.readAsDataURL(b);
                await new Promise(res=>r2.onloadend=res);
                  return r2.result.split(',')[1];
                  }
                  record();
                  ''')
```

```python
with open("voice.wav","wb") as f:
      f.write(b64decode(audio))
```
