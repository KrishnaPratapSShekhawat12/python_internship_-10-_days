# Conversions

## From Kg to pounds

```python
user_input=float(input("Enter the value to convert ="))
one_kg_value=2.20462
print(f"The value in pound is : {user_input*one_kg_value} lbs")
```

### From inches to Centimeter

```python
data=float(input("Enter the value to convert"))
one_inch_val=2.54
print(f"The value in centimeter is : {data*one_inch_val} cms")
```

### From fahrenheit to celsius

```python
data2=float(input("Enter the value to convert"))
print(f"The value in celsius is : {(data2-32)*(5/9)} degree celsius ")
```

### From Hours to minutes

```python
data3=float(input("Enter the value"))
print(f"The value in minutes is : {int(data3*60)} minutes")
```

### From Km/Hr to m/s

```python
data4=float(input("Enter the value to convert"))
print(f"The value in meter per second is : {data4*(5/18)} m/s")
```
